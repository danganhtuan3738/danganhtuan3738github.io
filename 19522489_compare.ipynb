```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

#%matplotlib inline
img1 = cv.imread('/content/drive/MyDrive/CS231.L21.KHTN/Sosanh2Anh/lena.jpg',0)

hist1,bins = np.histogram(img1.flatten(),256,[0,256])

cdf = hist1.cumsum()

cdf_normalized = cdf*float(hist1.max())/cdf.max()

plt.plot(cdf_normalized, color = 'b')

plt.hist(img1.flatten(),256,[0,256,],color = 'r')

plt.xlim([0,256])

plt.legend(('cdf','histogram'), loc = 'upper left')

plt.show()

```
