```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
#content images
img_raw= cv.imread('/content/drive/MyDrive/CS231.L21.KHTN/Tichchap/RawData.jpg',0)
img_lena=cv.imread('/content/drive/MyDrive/CS231.L21.KHTN/Tichchap/Lenna256x256.bmp',0)
img_ventg=cv.imread('/content/drive/MyDrive/CS231.L21.KHTN/Tichchap/ventg.bmp',0).astype(np.float32)
img_part=cv.imread('/content/drive/MyDrive/CS231.L21.KHTN/Tichchap/particles.bmp',0).astype(np.float32)
img_barbara=cv.imread('/content/drive/MyDrive/CS231.L21.KHTN/Tichchap/barbara.jpg',1)
```

```python
#Kernel
Kernel_smo=np.array([[1/9,1/9,1/9],
                    [1/9,1/9,1/9],
                    [1/9,1/9,1/9]])
Kernel_sharpen=np.array([[0,-1,0],
                      [-1,5,-1],
                      [0,-1,0]])
Kernel_direc=np.array([[0,1/4,0],
                      [0,2/4,0],
                      [0,-1/4,0]])
Kernel_ed_v=np.array([[-1,0,1],
                    [-1,0,1],
                    [-1,0,1]])
Kernel_ed_h=np.array([[-1,-1,-1],
                      [0,0,0],
                      [1,1,1]])
```

```python
#smoothing
img_rst_sm=cv.filter2D(img_raw,-1,Kernel_smo)
cv2_imshow(img_raw)
cv2_imshow(img_rst_sm)
```

```python
#Sharpening
img_rst_sh_l=cv.filter2D(img_lena,-1,Kernel_sharpen)
img_rst_sh_b=cv.filter2D(img_barbara,-1,Kernel_sharpen)
cv2_imshow(img_lena)
cv2_imshow(img_rst_sh_l)
cv2_imshow(img_barbara)
cv2_imshow(img_rst_sh_b)
```

```python
#Directional
img_rst_di=cv.filter2D(img_ventg,-1,Kernel_direc)
img_rst_di_transp=cv.filter2D(img_ventg,-1,np.transpose(Kernel_direc))
img_rst_re_di=np.sqrt(img_rst_di**2+img_rst_di_transp**2)
cv2_imshow(img_ventg)
cv2_imshow(img_rst_re_di)
```

```python
#Edge
img_rst_ed_v=cv.filter2D(img_part,-1,Kernel_ed_v)
img_rst_ed_h=cv.filter2D(img_part,-1,Kernel_ed_h)
img_rst_re_ed=np.sqrt(img_rst_ed_v**2 + img_rst_ed_h**2)
cv2_imshow(img_part)
cv2_imshow(img_rst_re_ed)
```
