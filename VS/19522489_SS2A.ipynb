```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from matplotlib import pyplot as plt
import math
from google.colab.patches import cv2_imshow
import os
```

```python
#peper
img_p = cv.imread('/content/drive/MyDrive/CS231.L21.KHTN/EdgeDetection/peppers.jpg',0).astype(np.float32)
#kernels
kernels_p_v=np.array([[-1,0,1],
                      [-2,0,2],
                      [-1,0,1]])
kernels_p_h=np.array([[-1,-2,-1],
                      [0,0,0],
                      [1,2,1]])
#img_rst
img_rst_p_v= cv.filter2D(img_p,-1,kernels_p_v)
img_rst_p_h= cv.filter2D(img_p,-1,kernels_p_h)
#show
cv2_imshow(np.sqrt(img_rst_p_v**2+img_rst_p_h**2))

```

```python
#particles
img_pa = cv.imread('/content/drive/MyDrive/CS231.L21.KHTN/EdgeDetection/particles.bmp',0).astype(np.float32)
#kernels
kernels_pa_v=np.array([[-1,0,1],
                       [-2,0,2],
                       [-1,0,1]])
kernels_pa_h=np.array([[-1,-2,-1],
                       [0,0,0],
                       [1,2,1]])
kernels_pa_cfl=np.array([[0,1,-1],
                        [-1,0,2],
                        [1,-2,0]])
kernels_pa_cfr=np.array([[-1,2,0],
                        [1,0,-2],
                        [0,-1,1]])
#img_rst
img_rst_pa_v= cv.filter2D(img_pa,-1,kernels_pa_v)
img_rst_pa_h= cv.filter2D(img_pa,-1,kernels_pa_h)
img_rst_pa_cfl= cv.filter2D(img_pa,-1,kernels_pa_cfl)
img_rst_pa_cfr= cv.filter2D(img_pa,-1,kernels_pa_cfr)
#show
cv2_imshow(np.sqrt(img_rst_pa_v**2+img_rst_pa_h**2+img_rst_pa_cfl**2+img_rst_pa_cfr**2))


```
